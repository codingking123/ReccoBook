# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
# books and its number of ratings
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Books and its avg rating
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# popular books, its number of ratings, and avg rating
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# this will  give only top 50 books which have minimun rating is by 250 and avg rating in descennding order
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

# Collaborative Filtering Based Recommender System

```python
# # we will see userID and how many books they have rated
# ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
# There are only 92106 users who have ever rated a book
# We will only take those users who have rated more than 200 books
# x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >20
# good_readers = x[x].index
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 5
good_readers = x[x].index
```

```python
# Here basically we are saying that among 10 lakh users(i.e.rating_with_name) only show those users who have rated more than 200 books
# filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_readers)]
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_readers)]
```

```python
filtered_rating
```

```python
# Similarly , here we will take only those books which has rated by more than 50 users, so that it'll be a popular
# y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
# famous_books = y[y].index
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=10
famous_books = y[y].index
```

```python
famous_books
```

```python
# final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
# final_ratings
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# final_ratings=final_ratings['Book-Title'].str.lower()
```

```python
final_ratings
```

```python
# popular_df.set_index(['Book-Title']).unstack()
# table = pd.pivot_table(popular_df, index=['Book-Title'],columns=['User-ID','Book-Rating'],aggfunc=np.sum)
```

```python
# we will create a pivot table
# pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
# pt.fillna(0,inplace=True)
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# We will treat every book as vector, and will calculate the euclidian distance bet them
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
# This shows we have calculated euclidian distance of 706 books with other 706 books
similarity_scores.shape
```

```python
# This represents the distance between book 1984 to the other books
similarity_scores[0]
```

```python
# def recommend(book_name):
#     # index fetch
#     book_name=book_name.lower()
#     try:
#         if((book_name == books['Book-Title']).any()):
#             index = np.where(pt.index==book_name)[0][0]
#             similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
#             for i in similar_items:
#                 print(pt.index[i[0]])
#     except:
#         pass
#     else:
#         if((book_name != books['Book-Title']).all()):
#             print("No book found")


def recommend(book_name):
    # index fetch
    
    book_name=book_name.lower()
    try:
        if((book_name == books['Book-Title']).any()):            
            index = np.where(pt.index==book_name)[0][0]
            similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:10]
    
            data = []
            for i in similar_items:
                item = []
                temp_df = books[books['Book-Title'] == pt.index[i[0]]]
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
                item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
                
                data.append(item)
    except:
        pass
    else:
        if((book_name != books['Book-Title']).all()):
            print("No book found")
    
    return data
```

```python
recommend('rich Dad, Poor Dad: What the Rich Teach Their Kids About Money--That the Poor and Middle Class Do Not!')
```
